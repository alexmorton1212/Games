```python
import requests
from bs4 import BeautifulSoup
```

```python
LETTER = "s"

WEBSITE_URL = "https://www.merriam-webster.com/wordfinder/classic/begins/common/5/" + LETTER + "/1"
WEBSITE_REQUEST = requests.get(WEBSITE_URL)

raw_soup = BeautifulSoup(requests.get(WEBSITE_URL).content, 'html.parser')

a_tags = raw_soup.find_all("a")
for a_tag in a_tags:
    text = a_tag.get_text()
    print(text)
```
